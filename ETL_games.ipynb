```python
import pandas as pd
import numpy as np
import ast
from dateutil.parser import parse
```

```python
games = pd.read_csv('DataBases/games.csv')
```

```python
games.info
```

```python
#Dropeamos los nulos de la columna 'id'
games = games.dropna(subset='id')
```

```python
games.dtypes
```

```python
#Definimos el formato de la columna generos como string.
games['genres'] = games['genres'].astype(str)
```

```python
#Formateamos, extraemos y generamos valores estilo booleanos para los generos.
games['genres'] = games['genres'].str.replace(r'\[|\]', '', regex=True)
games['genres'] = games['genres'].str.replace(']','',regex=True).replace("'","",regex=True)
games['genres'] = games['genres'].str.replace(';',',',regex=True)
genres_df = games['genres'].str.get_dummies(', ')
genres_df.columns = [col.capitalize() for col in genres_df.columns]
games = pd.concat([games, genres_df], axis=1)
```

```python
#Formateamos la columna 'release_date' y extraemos el anio de lanzamiento.
games["release_date"] = pd.to_datetime(games["release_date"], errors='coerce')
games["release_date"] = games["release_date"].apply(lambda x: parse(str(x)).strftime('%Y-%m-%d') if pd.notnull(x) else None)
games["release_date"] = pd.to_datetime(games["release_date"], errors='coerce')
games["release_year"] = games["release_date"].dt.year
```

```python
#Como vimos que tenemos fechas nulas, las dropeamos para alivianar el dataframe ya que no nos serviran esos registros.
games.dropna(subset=["release_date"], inplace=True)
games.drop_duplicates(subset=['id'])
```

```python
#Reemplazamos los precios como "free to play" por 0 y los "Starting at $449.00" por el valor correspondiente.
games['price'] = games['price'].replace(['Free To Play','Install Now','Third-party','Install Theme','Play Now','Free Movie','Free to Try', 'Play WARMACHINE: Tactics Demo','Play the Demo', 'Free','Free to Use', 'Free HITMAN™ Holiday Pack','Free to Play','Play for Free!','Free Mod','Free Demo', 'nan'], '0')
games['price'] = games['price'].replace(['Starting at $449.00'], 449.00)
games['price'] = games['price'].replace(['Starting at $499.00'], 499.00)
```

```python
#Imprimimos los valores unicos de la columna 'price'
unique_prices = games['price'].unique()
print(unique_prices)
```

```python
#Formateamos la columna 'price'
games['price'] = pd.to_numeric(games['price'], errors='coerce')
```

```python
#Rellenamos los nulos con '0'
games['price'].fillna(0, inplace=True)
```

```python
#Dropeamos todas las columnas que no nos hacen falta
games.drop('title', axis=1, inplace=True)
games.drop('price', axis=1, inplace=True)
games.drop('early_access', axis=1, inplace=True)
games.drop('genres', axis=1, inplace=True)
games.drop('url', axis=1, inplace=True)
games.drop('reviews_url', axis=1, inplace=True)
games.drop('release_date', axis=1, inplace=True)
games.drop('publisher', axis=1, inplace=True)
```

```python
#Ordenamos el dataframe en base a los 'id' de cada juego.
games = games.sort_values(by='id')
```

```python
#Y como ordenamos nuevamente el dataframe reiniciamos el indice
# a la vez que eliminamos el indice anterior
games.reset_index(inplace=True)
games.drop('index', axis=1, inplace=True)
```

```python
games.head()
```

```python
# Transformamos 'id' a int.
games['id'] = games['id'].astype(int)

# Transformamos 'release_year' a int.
games['release_year'] = games['release_year'].astype(int)
```

```python
games.head()
```

```python
#Volvemos a dropear columnas que no nos hacen falta
games.drop('tags', axis=1, inplace=True)
games.drop('specs', axis=1, inplace=True)
games.drop('Nan', axis=1, inplace=True)
```

```python
games.head()
```

```python
#Guardamos las transformaciones hechas en formato '.csv'.
games.to_parquet('DataBases/games.parquet', index=False)
```
